```python
# Read all four files into a Jupyter Notebook
import pandas as pd
import numpy as np
import seaborn as sns

book_sales_april = pd.read_excel(r"C:\Users\TAJ\Desktop\Python Data Prep and ETA\Data\Book_Sales_April.xlsx")
book_sales_may = pd.read_excel(r"C:\Users\TAJ\Desktop\Python Data Prep and ETA\Data\Book_Sales_May.xlsx")
book_sales_june = pd.read_excel(r"C:\Users\TAJ\Desktop\Python Data Prep and ETA\Data\Book_Sales_June.xlsx")

customers_data = pd.read_csv(r"C:\Users\TAJ\Desktop\Python Data Prep and ETA\Data\Book_Customers.csv")
```

## 2. Prepare Rows for Modeling

Hi again,

We are trying to predict which customers will purchase a book this month.

Can you reformat the data you compiled earlier this week so that it’s ready to be input into a model, with each row representing a customer instead of a purchase?

```python
# What should one row of data look like for the model?
df.head()
```

```python
book_sales_june.head()
```

```python
# Create a column called 'June Purchases' that sums up all purchases in June
june_purchases = (book_sales_june.groupby('Customer ID')['Book'].count()
.rename('June Purchases').to_frame()
.reset_index())

june_purchases.head()
```

```python
df.head()
```

```python
df_april_may = df[df['Purchase Date'].dt.month < 6]
df_april_may.tail()
```

```python
# Create a column called 'Total Spend' that sums up the prices of all the books that were purchased in April and May
total_spend = (df_april_may.groupby('Customer ID')['Price'].agg('sum')
              .rename('Total Spend')
                .to_frame()
                .reset_index())

total_spend.head()
```

```python
# Combine the 'June Purchases' and 'Total Spend' columns into a single DataFrame for modeling
model_df = total_spend.merge(june_purchases, how='left', on='Customer ID').fillna(0)
model_df.head()
```

## 3. Prepare Columns for Modeling

Hi again,

Thanks for your help earlier this week! We just learned that we also have to make all of the data numeric before inputting it into a predictive model.

Can you turn the Audience text field into a numeric field?

Thanks!\

```python
# Change the Audience field into a dummy variable
df.Audience.value_counts()
```

```python
audience_dummies = pd.get_dummies(df_april_may.Audience)
audience_dummies = audience_dummies.replace({True: 1, False: 0}).infer_objects(copy=False)
audience_dummies.head()
```

```python
df_april_may['Customer ID'].head()
```

```python
customer_audience = pd.concat([df_april_may['Customer ID'], audience_dummies], axis=1)
customer_audience.head()
```

```python
# Using the Audience dummy variables, create three new columns that contain
# the number of Adult / Children / Teen books purchased by each customer
book_types = customer_audience.groupby('Customer ID').sum().reset_index()
book_types.head()
```

```python
# Combine the three new columns back with the customer-level data
model_df.head()
```

```python
model_df = model_df.merge(book_types, how='left')
```

```python
has_kids2 = (pd.get_dummies(customers_data[['Customer ID', 'Has School Aged Children']], drop_first=True)
              .rename(columns={'Has School Aged Children_Yes': 'Has Kids'}))
has_kids2['Has Kids'] = has_kids2['Has Kids'].astype(int)
has_kids2.head()
```

```python
# create another feature
model_df = model_df.merge(has_kids2, how='left', on='Customer ID')
model_df.head()
```

```python
# add it to the dataframe for modeling
location = pd.get_dummies(df_april_may['Purchase Location'])
location = location.astype(int)
location.head()
```

```python
location_df = pd.concat([df_april_may, location], axis=1)[['Customer ID','In Person','Online']]
location_df.head()
```

```python
location_group = location_df.groupby('Customer ID').sum().reset_index()
location_group.head()
```

```python
# calculate the percent in person purchase by customer
in_person_pct = (location_group['In Person'] / (location_group['In Person'] + location_group['Online'])*100).rename('Percent in Person')
in_person_pct.head()
```

```python
# add it on th model dataframe
model_df = pd.concat([model_df, in_person_pct], axis=1)
model_df.head()
```
